```python
from sklearn.datasets import load_iris
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
iris = load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
data = pd.DataFrame(X)
data['cat'] = y
```

```python
xr, xe, yr, ye = train_test_split(X, y, test_size=0.1)
```

```python
sns.pairplot(data, hue='cat')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(xr, yr)
```

```python
yp = log_reg.predict(xe)
```

```python
from sklearn.metrics import f1_score, confusion_matrix
```

```python
print(confusion_matrix(yp, ye))
```

```python
f1_score(yp, ye)
```
