```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
data1 = np.random.normal(0, 0.1, 1000)
data2 = np.random.normal(1, 0.4, 1000) + np.linspace(0, 1, 1000)
data3 = 2 + np.random.random(1000) * np.linspace(0, 5, 1000)
data4 = np.random.normal(3, 0.2, 1000) + 0.3 * np.sin(np.linspace(0, 20, 1000))
```

```python
data = np.stack([data1, data2, data3, data4]).transpose()
```

```python
df = pd.DataFrame(data, columns=['data1', 'data2', 'data3', 'data4'])
df.head()
```

```python
df.plot(title='linear plot')
```

```python
plt.plot(data)
plt.title("Linear plot")
plt.legend(['Data1', 'Data2', 'Data3', 'Data4'])
```

```python
df.plot(style=".")
```

```python
df.plot(style=".", x='data1', y ='data2', xlim=(-1.5, 1.5), ylim=(0, 3))
```

```python
df.plot(kind="hist", bins = 50, title = "histogram", alpha = 0.8)
```

```python
df.plot(kind="hist", bins = 50, title = "histogram", alpha = 0.8, cumulative = True, normed = True)
```

```python
df.plot(kind="box")
```

```python
fig, ax = plt.subplots(2, 2, figsize = (12, 12))
df.plot(ax = ax[0][0])
df.plot(kind="hist", ax = ax[0][1], alpha = 0.8, bins = 100)
df.plot(style = "o", ax = ax[1][0])
df.plot(kind="box", ax = ax[1][1])
```
