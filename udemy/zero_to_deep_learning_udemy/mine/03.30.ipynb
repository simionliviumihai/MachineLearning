```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/weight-height.csv")
```

```python
data.plot(kind="scatter", x="Height", y="Weight")
```

```python
def line(x, w=0, b=0):
    return x*w + b

x = np.linspace(55, 80, 100)
```

```python
yhat = line(x, w = 7, b = -300)
```

```python
data.plot(kind = "scatter", x = "Height", y = "Weight")
plt.plot(x, yhat, color = "red")
```

```python
def mean_sq_err(yreal, ypred):
    s = (yreal - ypred) ** 2
    return s.sum()
```

```python
X = data["Height"].values
y = data["Weight"].values
```

```python
ypred = line(X, 7, -300)
```

```python
mean_sq_err(y, ypred)
```
