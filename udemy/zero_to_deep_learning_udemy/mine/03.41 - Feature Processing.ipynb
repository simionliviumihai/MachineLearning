```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("../data/weight-height.csv")
df.head()
```

```python
pd.get_dummies(df['Gender'], prefix='Gender').head()
```

```python
df["Height (feet)"] = df["Height"] / 12
df["Weight (100lbs)"] = df["Weight"] / 100
```

```python
df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
df["Heighmms"] = mms.fit_transform(df[["Height"]])
df["Weightmms"] = mms.fit_transform(df[["Weight"]])
```
